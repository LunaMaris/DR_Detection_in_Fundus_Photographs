This script is used to get all preprocessed and augmented data into single train and test arrays.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FinalArrays(DataBase, annotations = False):
    '''
    This function puts all preprocessed training and testing arrays together in 
    two final train and test arrays and saves these arrays
    '''
    Datapath = 'C:/Users/lunam/Documents/1steMaster/Stage/Data/' + DataBase + '/Arrays/'
    
    # bring all train arrays together
    print('Stacking train image arrays...')
    arr1 = np.load(Datapath + 'train_images_norm.npy')
    arr2 = np.load(Datapath + 'train_images_90_norm.npy')
    arr3 = np.load(Datapath + 'train_images_180_norm.npy')
    arr4 = np.load(Datapath + 'train_images_270_norm.npy')
    arr5 = np.load(Datapath + 'train_images_flip0_norm.npy')
    arr6 = np.load(Datapath + 'train_images_flip1_norm.npy')
    
    total_train_array = np.vstack([arr1, arr2, arr3, arr4, arr5, arr6])
    print('Shape total train image array: {}'.format(total_train_array.shape))
    np.save(Datapath + 'train_images_Final.npy', total_train_array)
    del arr1, arr2, arr3, arr4, arr5, arr6, total_train_array
    
    # bring all test arrays together
    print('Stacking test image arrays...')
    arr1 = np.load(Datapath + 'test_images_norm.npy')
    arr2 = np.load(Datapath + 'test_images_90_norm.npy')
    arr3 = np.load(Datapath + 'test_images_180_norm.npy')
    arr4 = np.load(Datapath + 'test_images_270_norm.npy')
    arr5 = np.load(Datapath + 'test_images_flip0_norm.npy')
    arr6 = np.load(Datapath + 'test_images_flip1_norm.npy')
    
    total_test_array = np.vstack([arr1, arr2, arr3, arr4, arr5, arr6])
    print('Shape total test image array: {}'.format(total_test_array.shape))
    np.save(Datapath + 'test_images_Final.npy', total_test_array)
    del arr1, arr2, arr3, arr4, arr5, arr6, total_test_array
    
    if annotations:
        
        # bring all train annotations together
        print('Stacking train annotation arrays...')
        arr1 = np.load(Datapath + 'train_annotations.npy')
        arr2 = np.load(Datapath + 'train_annotations_90.npy')
        arr3 = np.load(Datapath + 'train_annotations_180.npy')
        arr4 = np.load(Datapath + 'train_annotations_270.npy')
        arr5 = np.load(Datapath + 'train_annotations_flip0.npy')
        arr6 = np.load(Datapath + 'train_annotations_flip1.npy')
    
        total_train_array = np.vstack([arr1, arr2, arr3, arr4, arr5, arr6])
        print('Shape total train annotation array: {}'.format(total_train_array.shape))
        np.save(Datapath + 'train_annotations_Final.npy', total_train_array)
        del arr1, arr2, arr3, arr4, arr5, arr6, total_train_array
        
        # bring all test arrays together
        print('Stacking test annotation arrays...')
        arr1 = np.load(Datapath + 'test_annotations.npy')
        arr2 = np.load(Datapath + 'test_annotations_90.npy')
        arr3 = np.load(Datapath + 'test_annotations_180.npy')
        arr4 = np.load(Datapath + 'test_annotations_270.npy')
        arr5 = np.load(Datapath + 'test_annotations_flip0.npy')
        arr6 = np.load(Datapath + 'test_annotations_flip1.npy')

        total_test_array = np.vstack([arr1, arr2, arr3, arr4, arr5, arr6])
        print('Shape total test annotation array: {}'.format(total_test_array.shape))
        np.save(Datapath + 'test_annotations_Final.npy', total_test_array)
        del arr1, arr2, arr3, arr4, arr5, arr6, total_test_array
        
    # in case there are no annotation arrays, the labels have to be multiplied
    else:
        
        # labels for training
        print('Stacking train label arrays...')
        train_labels = np.load(Datapath+ 'train_labels.npy')
        train_labels_Final = np.vstack([train_labels,train_labels,train_labels,train_labels,train_labels,train_labels])
        print('Shape total train label array: {}'.format(train_labels_Final.shape))
        del train_labels, train_labels_Final
        
        # labels for testing
        print('Stacking test label arrays...')
        test_labels = np.load(Datapath+ 'test_labels.npy')
        test_labels_Final = np.vstack([test_labels,test_labels,test_labels,test_labels,test_labels,test_labels])
        print('Shape total test label array: {}'.format(test_labels_Final.shape))
        del test_labels, test_labels_Final
        
    print('Done')
```

```python
def PlotImage(image_array, database = '', image_name = '', save = False, savepath = ''):
    '''Visualize (and save if desired) an image represented by a numpy array'''
    
    plt.figure()
    plt.title(database+': '+image_name)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.imshow(image_array, cmap=plt.cm.Greys_r)
    
    if save == True:
        plt.savefig(savepath+database+'_'+image_name+'.png')
    
    return 
```
